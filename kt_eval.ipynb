# KT Evaluation (BKT vs Elo)

This notebook compares knowledge-tracing AUC on a held-out replay.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
sys.path.insert(0, str(ROOT / "src"))

from curriculum import load_curriculum
from kt_eval import simulate, report
```

```python
curriculum = load_curriculum()
trajectories = simulate(curriculum, n_learners=30, steps=40, seed=42)
metrics = report(trajectories)
metrics
```

```python
print("n_events:", metrics["n_events"])
print("BKT (item effect) AUC:", round(metrics["bkt_item_effect"]["auc"], 4))
print("Elo AUC:", round(metrics["elo"]["auc"], 4))
print("AUC gain (BKT - Elo):", round(metrics["bkt_item_effect"]["auc"] - metrics["elo"]["auc"], 4))
```
